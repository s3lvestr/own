1. Падіння з 3-го екрану (інтереси) до 4-го (електронна пошта)
Гіпотези:
Надмірна складність вибору інтересів:
Користувачам складно зрозуміти або вибрати інтереси.
Список інтересів занадто довгий або непривабливий.

Недостатня мотивація продовжувати:
Користувачі не розуміють, як вибір інтересів впливає на їхній досвід у додатку.

Проблеми з UX/UI:
Інтерфейс екрану виглядає перевантаженим, нелогічним або складним для використання.


Варіанти вирішення:
Оптимізувати вибір інтересів:
Зменшити кількість інтересів або автоматично пропонувати найпопулярніші варіанти.
Додати кнопку «Пропустити» або «Вибрати пізніше».

Мотивувати користувачів:
Додати пояснення, як вибір інтересів покращить їхній досвід.
Використовувати мікроанімовані підказки чи прогрес-бар для заохочення.

Покращити UX/UI:
Спрощений, мінімалістичний дизайн.
Використання іконок, щоб зробити вибір інтересів інтуїтивно зрозумілим.
Тестування за допомогою налаштованих тригерів у гугл аналітиці, або використання Hotjar.
